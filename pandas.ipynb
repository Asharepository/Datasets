```python
scores = [[90,67,89,76,93,90],
          [93,80,75,86,88,92],
          [82,72,67,91,87,79],
          [78,85,77,98,90,72],
          [84,87,91,85,75,82],
          [91,83,90,76,78,93]]
```

```python
pd.DataFrame(scores)
```

```python
ind = pd.MultiIndex.from_product([['school1','school2','school3'],[2019,2020]])
ind
```

```python
cols = pd.MultiIndex.from_product([['7th grade','8th grade'],['Math','Reading','Writing']])
cols
```

```python
df = pd.DataFrame(scores,index = ind, columns = cols)
df
```

```python
df.index.names = ['Schools','Years']
df.columns.names = ['Grades','Subjects']
df
```

```python
import pandas as pd

# Provide the raw URL of the CSV file
url = 'https://raw.githubusercontent.com/Asharepository/Datasets/master/StudentsPerformance.csv'

# Read the CSV file
df = pd.read_csv(url)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
df.index.names
```

```python
mdf = df.set_index(['race/ethnicity','parental level of education','lunch','test preparation course','gender'])
mdf.head()
```

```python
mdf.index.names
```

```python
mdf.loc['group A']
```

```python
mdf.loc['group A','some college','standard','completed','male']
```

```python
mdf.loc['group B','completed'] # you cannot slip in between like this if you want to slip the index values then use slice None
```

```python
mdf.loc['group B',slice(None),slice(None),'completed']
```

```python
mdf.loc[['group B','group A'],['high school','some college'],slice(None),'completed']
```
